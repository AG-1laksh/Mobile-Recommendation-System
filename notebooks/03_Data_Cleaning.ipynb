# Data Cleaning

## Objective

The objective of this notebook is to clean and standardize the merged Samsung smartphone dataset before Exploratory Data Analysis (EDA).

The cleaning process includes:

- Missing value analysis
- Duplicate detection
- Data type conversion
- Text standardization
- Feature extraction
- Final validation

Output:

data/processed/cleaned_dataset.csv

```python
import pandas as pd
import numpy as np
import re

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("../data/interim/SamsungPhoneData_Updated.csv")

print("Dataset Loaded Successfully")
```

```python
print("Shape :", df.shape)

df.head()
```

```python
df.info()
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

missing.sort_values(
    by="Missing Values",
    ascending=False
)
```

```python
duplicates = df.duplicated().sum()

print(f"Duplicate Rows : {duplicates}")
```

```python
df.dtypes
```

```python
yes_no_columns = [
    "OIS",
    "Waterproof"
]

for col in yes_no_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )
```

```python
df["AI_Features"] = (
    df["AI_Features"]
    .astype(str)
    .str.strip()
    .replace({
        "None":"No",
        "Nan":"No"
    })
)
```

```python
camera_columns = [
    "UltraWide_MP",
    "Telephoto_MP"
]

df[camera_columns] = (
    df[camera_columns]
    .fillna(0)
)
```

```python
df.isnull().sum()
```

```python
# Save cleaned dataset
df.to_csv(
    "../data/processed/cleaned_dataset.csv",
    index=False
)

print("✅ Cleaned dataset saved successfully!")
```
